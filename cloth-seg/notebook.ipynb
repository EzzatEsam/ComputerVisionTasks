```python
cat_id_dict = dict(zip(categories_df["id"], categories_df["name"]))
cat_supercat_dict = dict(zip(categories_df["id"], categories_df["supercategory"]))
```

```python
annotation_df["category_name"] = annotation_df["category_id"].map(cat_id_dict)
annotation_df["supercategory"] = annotation_df["category_id"].map(cat_supercat_dict)
annotation_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)
```

# Annotation Statistics Visualization

Let's explore the distribution and statistics of the annotations in the dataset.

```python
# 1. Distribution of annotations by category
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Category distribution (bar plot)
category_counts = annotation_df['category_name'].value_counts()
axes[0, 0].barh(category_counts.index[:20], category_counts.values[:20], color='steelblue')
axes[0, 0].set_xlabel('Count', fontsize=12)
axes[0, 0].set_title('Top 20 Categories by Annotation Count', fontsize=14, fontweight='bold')
axes[0, 0].invert_yaxis()

# Supercategory distribution (pie chart)
supercategory_counts = annotation_df['supercategory'].value_counts()
colors = plt.cm.Set3(range(len(supercategory_counts)))
axes[0, 1].pie(supercategory_counts.values, labels=supercategory_counts.index, 
               autopct='%1.1f%%', startangle=90, colors=colors)
axes[0, 1].set_title('Distribution by Supercategory', fontsize=14, fontweight='bold')

# Category distribution by supercategory (stacked bar)
category_super_df = annotation_df.groupby(['supercategory', 'category_name']).size().unstack(fill_value=0)
category_super_df.plot(kind='bar', stacked=True, ax=axes[1, 0], legend=False)
axes[1, 0].set_xlabel('Supercategory', fontsize=12)
axes[1, 0].set_ylabel('Count', fontsize=12)
axes[1, 0].set_title('Category Distribution within Supercategories', fontsize=14, fontweight='bold')
axes[1, 0].tick_params(axis='x', rotation=45)

# Top 15 categories
top_categories = annotation_df['category_name'].value_counts().head(15)
axes[1, 1].bar(range(len(top_categories)), top_categories.values, color='coral')
axes[1, 1].set_xticks(range(len(top_categories)))
axes[1, 1].set_xticklabels(top_categories.index, rotation=45, ha='right')
axes[1, 1].set_ylabel('Count', fontsize=12)
axes[1, 1].set_title('Top 15 Most Common Categories', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 2. Annotation area statistics
annotation_df['area'] = annotation_df['area'].astype(float)

fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Area distribution (histogram)
axes[0, 0].hist(annotation_df['area'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Area (pixels)', fontsize=12)
axes[0, 0].set_ylabel('Frequency', fontsize=12)
axes[0, 0].set_title('Distribution of Annotation Areas', fontsize=14, fontweight='bold')
axes[0, 0].set_yscale('log')

# Area distribution (log scale)
axes[0, 1].hist(annotation_df['area'], bins=50, color='lightcoral', edgecolor='black', alpha=0.7)
axes[0, 1].set_xlabel('Area (pixels)', fontsize=12)
axes[0, 1].set_ylabel('Frequency', fontsize=12)
axes[0, 1].set_title('Distribution of Annotation Areas (Log-Log Scale)', fontsize=14, fontweight='bold')
axes[0, 1].set_xscale('log')
axes[0, 1].set_yscale('log')

# Box plot of area by supercategory
annotation_df.boxplot(column='area', by='supercategory', ax=axes[1, 0])
axes[1, 0].set_xlabel('Supercategory', fontsize=12)
axes[1, 0].set_ylabel('Area (pixels)', fontsize=12)
axes[1, 0].set_title('Area Distribution by Supercategory', fontsize=14, fontweight='bold')
axes[1, 0].set_yscale('log')
plt.sca(axes[1, 0])
plt.xticks(rotation=45, ha='right')

# Mean area by top categories
top_cats = annotation_df['category_name'].value_counts().head(15).index
mean_area_by_cat = annotation_df[annotation_df['category_name'].isin(top_cats)].groupby('category_name')['area'].mean().sort_values(ascending=False)
axes[1, 1].barh(mean_area_by_cat.index, mean_area_by_cat.values, color='mediumseagreen')
axes[1, 1].set_xlabel('Mean Area (pixels)', fontsize=12)
axes[1, 1].set_title('Mean Area by Top 15 Categories', fontsize=14, fontweight='bold')
axes[1, 1].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# 3. Annotations per image statistics
annotations_per_image = annotation_df.groupby('image_id').size()

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Distribution of annotations per image
axes[0].hist(annotations_per_image, bins=30, color='plum', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Number of Annotations', fontsize=12)
axes[0].set_ylabel('Number of Images', fontsize=12)
axes[0].set_title('Distribution of Annotations per Image', fontsize=14, fontweight='bold')
axes[0].axvline(annotations_per_image.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {annotations_per_image.mean():.2f}')
axes[0].axvline(annotations_per_image.median(), color='green', linestyle='--', linewidth=2, label=f'Median: {annotations_per_image.median():.2f}')
axes[0].legend()

# Box plot
axes[1].boxplot(annotations_per_image, vert=True)
axes[1].set_ylabel('Number of Annotations', fontsize=12)
axes[1].set_title('Annotations per Image (Box Plot)', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Summary statistics
stats_text = f"""
Total Images: {len(annotations_per_image)}
Total Annotations: {len(annotation_df)}
Mean Annotations/Image: {annotations_per_image.mean():.2f}
Median Annotations/Image: {annotations_per_image.median():.2f}
Min Annotations/Image: {annotations_per_image.min()}
Max Annotations/Image: {annotations_per_image.max()}
Std Dev: {annotations_per_image.std():.2f}
"""
axes[2].text(0.1, 0.5, stats_text, fontsize=12, verticalalignment='center', 
             bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
axes[2].axis('off')
axes[2].set_title('Summary Statistics', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# 4. Category statistics summary table
print("="*80)
print("CATEGORY STATISTICS SUMMARY".center(80))
print("="*80)

category_stats = annotation_df.groupby('category_name').agg({
    'id': 'count',
    'area': ['mean', 'median', 'std', 'min', 'max']
}).round(2)

category_stats.columns = ['Count', 'Mean Area', 'Median Area', 'Std Area', 'Min Area', 'Max Area']
category_stats = category_stats.sort_values('Count', ascending=False)

print("\nTop 20 Categories by Count:")
print(category_stats.head(20).to_string())

print("\n" + "="*80)
print("\nSupercategory Statistics:")
supercategory_stats = annotation_df.groupby('supercategory').agg({
    'id': 'count',
    'area': ['mean', 'median']
}).round(2)
supercategory_stats.columns = ['Count', 'Mean Area', 'Median Area']
print(supercategory_stats.to_string())
```
